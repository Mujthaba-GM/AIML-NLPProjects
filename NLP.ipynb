```python
#Removing Whitespace
text = "   Hello World   "
print(text.strip())   # 'Hello World'
print(text.lstrip())  # 'Hello World   '
print(text.rstrip())  # '   Hello World'
```

```python
 #Removing Punctuation & Special Characters
import re
text = "Hello!!! Are you #1?"
text = re.sub(r'[^\w\s]', '', text)
print(text)  # Hello Are you 1
```

```python
 #Removing Numbers
 text = "I have 2 cats and 3 dogs"
text = re.sub(r'\d+', '', text)
print(text)  # I have  cats and  dogs
```

```python
#Whitespace Normalization
text = "This    has   extra   spaces"
text = " ".join(text.split())
print(text)  # This has extra spaces
```

```python
#Tokenizations
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp(text)

tokens = [token.text for token in doc]
print("Step  Tokens:\n", tokens)
```

```python
# Step  — Remove Stopwords
tokens_no_stop = [token.text for token in doc if not token.is_stop]
print("Step 7 Tokens without Stopwords:\n", tokens_no_stop)
```

```python
#Step — Lemmatization
lemmas = [token.lemma_ for token in doc if not token.is_stop]
print("Step 8️⃣ Lemmas:\n", lemmas)
```

```python
#Step 🔟 — Spelling Correction
from textblob import TextBlob

corrected = str(TextBlob(text).correct())
print("Step 🔟 Spelling Corrected Text:\n", corrected)

```

```python
#POS taggging
pos_tags = [(token.text, token.pos_) for token in doc]
print("Step 🔟 POS Tags:\n", pos_tags)
```

```python
# Named Entity Recognition (NER)
entities = [(ent.text, ent.label_) for ent in doc.ents]
print("Named Entities:", entities)

# Remove named entities from text
tokens_no_ents = [token.text for token in doc if not token.ent_type_]
print("Text without entities:", tokens_no_ents)
```

```python
#Remove Non-Alphabetic Tokens
tokens_alpha = [token.text for token in doc if token.is_alpha]
print("Alphabetic Tokens:", tokens_alpha)
```

```python
 #Remove Short Words
 tokens_long = [token.text for token in doc if len(token.text) >= 3]
print("Tokens ≥ 3 characters:", tokens_long)
```

```python
 #Remove Pronouns (or Specific POS)
 tokens_no_pron = [token.text for token in doc if token.pos_ != 'PRON']
print("Tokens without pronouns:", tokens_no_pron)
```

```python
#scentence segmentations
sentences = [sent.text.strip() for sent in doc.sents]
print("Sentences:", sentences)
```

```python
 #Custom Stopwords
 from spacy.lang.en.stop_words import STOP_WORDS

custom_stopwords = STOP_WORDS.union({'intelligence', 'language'})
tokens_no_custom_stop = [token.text for token in doc if token.text not in custom_stopwords]
print("Tokens without custom stopwords:", tokens_no_custom_stop)
```
